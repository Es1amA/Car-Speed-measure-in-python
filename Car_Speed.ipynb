```python
import cv2
import time
```

```python
car_cascade = cv2.CascadeClassifier('cars.xml') 
```

```python

#polygon 
coord=[[50,200],[700,200],[50,500],[700,500]]
```

```python
cap = cv2.VideoCapture('traffic.mp4') 

#Distance between two horizontal lines in (meter)
dist = 3
while True:
    ret, frame = cap.read()
    if not ret:
        #print("...")
        break
    dim = (1000, 1000)
    #print("...")
    resized = cv2.resize(frame, dim)
    gray = cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
    cars=car_cascade.detectMultiScale(gray,1.8,2)
    for (x,y,w,h) in cars:
        cv2.rectangle(resized,(x,y),(x+w,y+h),(225,0,0),2)
        cv2.line(resized, (coord[0][0],coord[0][1]),(coord[1][0],coord[1][1]),(0,0,255),2)   #First horizontal line
        cv2.line(resized, (coord[0][0],coord[0][1]), (coord[2][0],coord[2][1]), (0, 0, 255), 2) #Vertical left line
        cv2.line(resized, (coord[2][0],coord[2][1]), (coord[3][0], coord[3][1]), (0, 0, 255), 2) #Second horizontal line
        cv2.line(resized, (coord[1][0],coord[1][1]), (coord[3][0], coord[3][1]), (0, 0, 255), 2) #Vertical right line

    for (x, y, w, h) in cars:

        if(x>coord[0][0] and x<coord[1][0] and y>coord[0][1] and y<coord[2][1] ):
            cv2.line(resized, (coord[0][0], coord[0][1]), (coord[1][0], coord[1][1]), (0, 255,0), 2) #Changes line color to green
            global tim1
            time1= time.time() #Initial time
            now = time.ctime(int(time1))
            print("Car Entered at: ",str(now))
        if (x>coord[3][0] and y>coord[3][1]):
            cv2.line(resized, (coord[2][0],coord[2][1]), (coord[3][0], coord[3][1]), (0, 0, 255), 2) #Changes line color to green
            global time2
            time2 = time.time() #Final time
            now = time.ctime(int(time2))
            print("Car Left at: ",now)
            #We know that distance is 3m\

            div=abs(time1-time2)
            if div !=0:
                print("Speed in (m/s) is:", dist / ((time2 - time1)))

    cv2.imshow('img',resized) #Shows the frame
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
